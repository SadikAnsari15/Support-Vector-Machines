```python
# SVM Classification
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV     # automated hyperparameter tunning
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import metrics
from sklearn.model_selection import train_test_split, cross_val_score

```

```python
data1=pd.read_csv("E:\\materials\\Assignment\\Assignment17\\SalaryData_Test(1).csv")
data2=pd.read_csv("E:\\materials\\Assignment\\Assignment17\\SalaryData_Train(1).csv")
```

```python
test = data1.copy()
train = data2.copy()
```

```python
test.head()
```

```python
train.head()
```

```python
lst = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
for i in lst:
    train[i]= label.fit_transform(train[i])
  
```

```python
  for i in lst:
    test[i]= label.fit_transform(test[i])
```

```python
train.head()
```

```python
x_train = train.iloc[:,0:102]
y_train = train['Salary']
```

```python
x_test = test.iloc[:,0:102]
y_test = test['Salary']
```

```python
x_test1 = x_test.drop(['Salary'],axis = 1)
```

```python
x_train1 = x_train.drop(['Salary'], axis = 1)
```

```python
x_train
```

```python
# SVM with kernel rbf
clf = SVC(kernel='rbf')
clf.fit(x_train1 , y_train)
y_pred = clf.predict(x_test1)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
#  SVM with kernel poly
clf = SVC(kernel='poly')
clf.fit(x_train1 , y_train)
y_pred = clf.predict(x_test1)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
#  SVM with kernel sigmoid
clf = SVC(kernel='sigmoid')
clf.fit(x_train1 , y_train)
y_pred = clf.predict(x_test1)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
